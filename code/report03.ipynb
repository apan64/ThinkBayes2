## Allergic Guests Problem

Suppose you are having a dinner party with 10 guests and 4 of them are allergic to cats. Because you have cats, you expect 50% of the allergic guests to sneeze during dinner. At the same time, you expect 10% of the non-allergic guests to sneeze. What is the distribution of the total number of guests who sneeze?

### Answer

The first thing to notice about this problem is that their are two groups of guests with separate probabilities of sneezing.  Since the problem asks for the distribution of the **total** number of guests who sneeze, we can actually find the distributions for the two individual groups then add them together.

To find the distributions, we can use the MakeBinomialPmf function which takes two arguments, n and p, and returns the distribution of successes with probability p in n attempts.  We first need to determine the values of n and p for the allergic and non-allergic groups in order to use MakeBinomialPmf.  Fortunately, the problem directly gives us the numbers we need: for the allergic group, there are 4 guests with a 50% probability of sneezing, and that leaves 6 non-allergic guests with a 10% probability of sneezing during dinner.

```python
allergic = 4
nonallergic = 6
allergic_sneeze = .5
nonallergic_sneeze = .1
```

Now all we have to do is plug those numbers into the MakeBinomialPmf function and plot it.

```python
from thinkbayes2 import MakeBinomialPmf
% matplotlib inline
import thinkplot

pmf = MakeBinomialPmf(allergic, allergic_sneeze) + MakeBinomialPmf(nonallergic, nonallergic_sneeze)
thinkplot.Hist(pmf, label = 'Sneeze Probability')
thinkplot.Config(xlabel = 'Sneezes', ylabel = 'PMF')
```

And there we have our distribution of guests who will sneeze during dinner!  We can also quickly find the mean of the distribution to know exactly where the distribution is centered about.

```python
pmf.Mean()
```

## Gluten Sensitivity Problem

[This study from 2015](http://onlinelibrary.wiley.com/doi/10.1111/apt.13372/full) showed that many subjects diagnosed with non-celiac gluten sensitivity (NCGS) were not able to distinguish gluten flour from non-gluten flour in a blind challenge.

Here is a description of the study:

>"We studied 35 non-CD subjects (31 females) that were on a gluten-free diet (GFD), in a double-blind challenge study. Participants were randomised to receive either gluten-containing ﬂour or gluten-free ﬂour for 10 days, followed by a 2-week washout period and were then crossed over. The main outcome measure was their ability to identify which ﬂour contained gluten.
>"The gluten-containing ﬂour was correctly identiﬁed by 12 participants (34%)..."
Since 12 out of 35 participants were able to identify the gluten flour, the authors conclude "Double-blind gluten challenge induces symptom recurrence in just one-third of patients fulﬁlling the clinical diagnostic criteria for non-coeliac gluten sensitivity."

This conclusion seems odd to me, because if none of the patients were sensitive to gluten, we would expect some of them to identify the gluten flour by chance.  So the results are consistent with the hypothesis that none of the subjects are actually gluten sensitive.

We can use a Bayesian approach to interpret the results more precisely.  But first we have to make some modeling decisions.

1. Of the 35 subjects, 12 identified the gluten flour based on resumption of symptoms while they were eating it.  Another 17 subjects wrongly identified the gluten-free flour based on their symptoms, and 6 subjects were unable to distinguish.  So each subject gave one of three responses.  To keep things simple I follow the authors of the study and lump together the second two groups; that is, I consider two groups: those who identified the gluten flour and those who did not.

2. I assume (1) people who are actually gluten sensitive have a 95% chance of correctly identifying gluten flour under the challenge conditions, and (2) subjects who are not gluten sensitive have only a 40% chance of identifying the gluten flour by chance (and a 60% chance of either choosing the other flour or failing to distinguish).

Using this model, estimate the number of study participants who are sensitive to gluten.  What is the most likely number?  What is the 95% credible interval?

### Answer

To estimate the number of participants who are actually gluten-sensitive, we need to create a probability distribution based on the given data.  To do this we'll create a class for glutenSensitivity, which will extend the Suite class.  We will modify the inital statement slightly so that it looks for two arguments: the usual hypotheses, and in this case the total amount of study participants.  The hypotheses will be passed onto the initial statement for Suite, while the total number of participants will be stored in the instance of the glutenSensitivity class.

Next, we have to define the Likelihood function.  The data passed into the Likelihood function is the number people who correctly identified the bag of gluten, while the hypothesis is the number of people who are actually gluten-sensitive.  Since we have to return the likelihood of the data given the hypothesis, we need to determine the probability of the number of participants identifying gluten for various hypotheses of people who are gluten-sensitive.  Similar to the last problem, there are actually two groups, this time based on the hypothesis: participants who are gluten-sensitive, and participants who are not gluten-sensitive.  Again, we can use the MakeBinomialPmf function for both groups and add the resulting distributions together to return the Pmf value for the number of people who correctly identified the gluten.

```python
from thinkbayes2 import Suite

class glutenSensitivity(Suite):
    def __init__(self, prior, total):
        Suite.__init__(self, prior)
        self.subjectTotal = total
        
    def Likelihood(self, data, hypo):
        '''
        Compute the likelihood of 'data' given 'hypo'.

        data: number of participants who correctly identified the gluten
        hypo: number of participants who are gluten-sensitive
        returns: likelihood of the data given the the hypothesis
        '''
        glutenPmf = MakeBinomialPmf(hypo, .95) + MakeBinomialPmf(self.subjectTotal - hypo, .4)
        return glutenPmf[data]
```

After constructing our class, we just have to create it with a range of hypotheses and the number of participants.  Because there was a group of 35 people, there can be any number of gluten-sensitive individuals within that group so we'll just use all numbers from 0 to 35 as our hypotheses.  Next, we update our class with the number of people who were correct and plot our class.  We can also use the CredibleInterval method to find the 95% credible interval.

```python
glutenDistribution = glutenSensitivity(range(0, 36), 35)
glutenDistribution.Update(12)
thinkplot.Pdf(glutenDistribution, label = 'Gluten Sensitive Probability')
thinkplot.Config(xlabel = 'Gluten Sensitive', ylabel = 'PMF')
print('The 95% credible interval is from: ' + str(glutenDistribution.CredibleInterval(95)))
```

And there we have it!  The most likely number of gluten-sensitive participants is 0, but all values up to 8 participants are within the 95% credible range.

## Kim Rhode Problem

At the 2016 Summer Olympics in the Women's Skeet event, Kim Rhode faced Wei Meng in the bronze medal match.  They each hit 15 of 25 skeets, sending the match into sudden death.  In the first round, both hit 1 of 2 skeets.  In the next two rounds, they each hit 2 skeets.  Finally, in the fourth round, Rhode hit 2 and Wei hit 1, so Rhode won the bronze medal, making her the first Summer Olympian to win an individual medal at six consecutive summer games.

But after all that shooting, what is the probability that Rhode is actually a better shooter than Wei?  If the same match were held again, what is the probability that Rhode would win?

As always, you will have to make some modeling decisions, but one approach is to estimate, for each shooter, the probability of hitting a skeet.  Then, to estimate the probability that Rhode is a better shooter, you can draw samples from the two posterior distributions and compare them.  To estimate the probability of winning a rematch, you could draw samples from the posterior distributions and simulate a round of 25 shots.

### Answer

To start with, the probability of skeet shooters having a uniform distribution of probabilities of hitting a skeet is inaccurate, but since we don't have a large amount of data the alpha and beta values should only be slightly changed to produce a curve in the distribution.  Next, we update each shooter's Beta class with the number of skeets they hit and missed.  

To determine the probability of Rhode being a better shooter than Wei, we can create a Cdf for a large number of sample data based off of the current probability distribution in the Beta classes.  Using numpy's mean function, we can compute the percentage of sample trials in which Rhode had a higher score than Wei.

To determine the probability that Rhode would win again in a rematch, we can use numpy's random.binomial function to create a set of skeet shots for Rhode and Wei based off of the Cdf data again, and count the number of times that Rhode's score is greater than Wei's.

```python
from thinkbayes2 import Pmf, Cdf, Suite, Beta, Joint
import numpy as np

rhode = Beta(2, 2)
rhode.Update((22, 11))
wei = Beta(2, 2)
wei.Update((21, 12))

rhode_sample = rhode.MakeCdf(10001).Sample(10000)
wei_sample = wei.MakeCdf(10001).Sample(10000)
print('Probability that Rhode is a better shooter: ' + str(np.mean(rhode_sample > wei_sample)))

rhode_rematch = np.random.binomial(25, rhode_sample)
wei_rematch = np.random.binomial(25, wei_sample)
count = 0
for p1, p2 in zip(rhode_rematch, wei_rematch):
    count += 1 if p1 > p2 else 0
print('Probability that Rhode would win a rematch: ' + str(count/len(rhode_rematch)))
```

### Answer

Given only the information on how many ties Rhode and Wei had and no information on their scores doesn't tell us too much about their accuracy.  

## Improving Reading Ability Problem

> An educator conducted an experiment to test whether new directed reading activities in the classroom will help elementary school pupils improve some aspects of their reading ability. She arranged for a third grade class of 21 students to follow these activities for an 8-week period. A control classroom of 23 third graders followed the same curriculum without the activities. At the end of the 8 weeks, all students took a Degree of Reading Power (DRP) test, which measures the aspects of reading ability that the treatment is designed to improve.

> Summary statistics on the two groups of children show that the average score of the treatment class was almost ten points higher than the average of the control class. A two-sample t-test is appropriate for testing whether this difference is statistically significant. The t-statistic is 2.31, which is significant at the .05 level.

Run this analysis again for the control group.  What is the distribution of the difference between the groups?  What is the probability that the average "reading power" for the treatment group is higher?  What is the probability that the variance of the treatment group is higher?

### Answer

To compare the differences between the control and treated groups, we have to first run both cases.

```python
# The code for the treated group
import pandas as pd
from scipy.stats import norm
from itertools import product

df = pd.read_csv('drp_scores.csv', skiprows=21, delimiter='\t')
grouped = df.groupby('Treatment')
for name, group in grouped:
    print(name, group.Response.mean())

class Normal(Suite, Joint):
    
    def Likelihood(self, data, hypo):
        """
        
        data: sequence of test scores
        hypo: mu, sigma
        """
        mu, sigma = hypo
        likes = norm.pdf(data, mu, sigma)
        return np.prod(likes)
    
mus = np.linspace(20, 80, 101)
sigmas = np.linspace(5, 30, 101)

control = Normal(product(mus, sigmas))
data = df[df.Treatment=='Control'].Response
control.Update(data)

pmf_mu0 = control.Marginal(0)

pmf_sigma0 = control.Marginal(1)
```

After creating a new instance of the Normal class and updating it with data from the csv, we can plot the resulting contour plot.

```python
treated = Normal(product(mus, sigmas))
data = df[df.Treatment=='Treated'].Response
treated.Update(data)
thinkplot.Contour(treated, pcolor=True)
thinkplot.Config(xlabel='mu', ylabel='Pmf')
```

Next, we can plot the marginal distributions for both mu and sigma.

```python
pmf_mu1 = treated.Marginal(0)
thinkplot.Pdf(pmf_mu1, label = 'mu')
thinkplot.Config(xlabel='mu', ylabel='Pmf')
```

```python
pmf_sigma1 = treated.Marginal(1)
thinkplot.Pdf(pmf_sigma1, label = 'sigma')
thinkplot.Config(xlabel='sigma', ylabel='Pmf')
```

Now we can determine the difference in the distributions between the two groups.  Since the marginal distributions are Beta classes, we can subtract them and take the mean of the difference for the average difference between the two, and use the MAP method to determine the most common difference.

```python
dif = pmf_mu1 - pmf_mu0
print('The mean of the distribution difference is: ' + str(dif.Mean()))
print('The value with maximum a posteori probability is: ' + str(dif.MAP()))
```

Using the ProbGreater method we can determine the probability that the mu values for the control group are greater.

```python
pmf_mu1.ProbGreater(pmf_mu0)
```

And the probability that the sigma values for the control group are greater.

```python
pmf_sigma1.ProbGreater(pmf_sigma0)
```

## Tinder Problem Part 1

On Tinder, a dating app where users can "swipe right" on people they are interested in meeting and "swipe left" on those they aren't interested in, Joe swipes right on the same number of people every day.  Over the course of a week, Joe matches with 2, 4, 3, 2, 6, 4, 5 people.  Based on this data, how many swipes did Joe make a day and at what rate did he match with people?

### Answer

In this problem we know how many people Joe matched with but we have no idea how many total right swipes he made or what the probability of matching was.  In order to solve this problem, we'll have to assume that Joe's love life isn't entirely hopeless, and that at the most he only made 100 swipes.  We know that Joe matched with 5 people at one point, so the minimum number of swipes he is making is 5.  Since we have two unknown parameters, we'll create a class that inherits from both Suite and Joint.  The Likelihood function for our class will simply return the binomial for the data given the number of swipes and probability of matching as hypothesis, since we're looking for the probability of the data (in this case, the probability of the given number of matches/successes in the total number of swipes/trials under the probability).

```python
from scipy.stats import binom
from itertools import product
from thinkbayes2 import Joint
import numpy as np

class Tinder(Suite, Joint):
    def Likelihood(self, data, hypo):
        """
        Compute the likelihood of 'data' given 'hypo'.

        data: number of matches
        hypo: tuple of the total number of swipes and the probability of matching

        returns: likelihood of the data given the probability of matching and the total number of swipes
        """
        swipes, matchProb = hypo
        return 0 if data > swipes else binom.pmf(data, swipes, matchProb)
```

```python
t = Tinder(product(range(5, 101), np.linspace(0, 1, 101)))
for matches in [2, 4, 3, 2, 6, 4, 5]:
    t.Update(matches)
thinkplot.Contour(t, pcolor = True)
thinkplot.Config(xlabel = 'Right Swipes', ylabel = 'Matching Probability')
```

With the information given, we can't really draw any firm conclusions about Joe's daily swipe count or the rate at which he matched with people.  However, we can somewhat tell by this contour plot, and by plotting a cross-section of the contour plot, that it is much more likely for Joe to have a low swipe total and a high match rate than it is for him to have a high swipe total and a low match rate.

```python
thinkplot.Pdf(t.Marginal(0))
thinkplot.Config(xlabel = 'Right Swipes', ylabel = 'PMF')
```

```python
thinkplot.Pdf(t.Marginal(1))
thinkplot.Config(xlabel = 'Matching Probability', ylabel = 'PMF')
```

## Tinder Problem Part 2

Joe believes that Tinder's new "super like" option yields three times as many matches as swiping right does.  After his last week of Tinder, Joe continues to swipe right on the same number of people, but also now super likes one person per day.  With this method, he matches with 8 people on one day, but only 2 and 1 on the next two days.  As Joe is grieving over his romantic rollercoaster, how does Joe's belief in the effectiveness of the super like affect the distribution of his swipes and his matching rate?

### Answer

With the introduction of the super like, we now have to redefine our class to handle having one super like per day.  This involves adding an additional return state in the event of a super like.  This extra return state takes a bit of thought: we need to account for the likelihoods of both matching from normal swipes and from super likes.  We'll need to create a new class to handle all of this information.

In our new class, the data argument for our Likelihood function will change to a tuple consisting of a number and a boolean.  The number will be the same as the old data argument, while the boolean will be True if there was a super like, and False otherwise.

The return statement will have the same parts of the old Tinder class with one additional condition: if there was a super like and three times the probability of matching is less than 1, we have to include the probability of success for a super like.  As in the allergy and gluten problems, this involves adding together multiple probabilities, so we can use the MakeBinomialPmf function to find the probability distribution for swipes and for the super like and add them together.  We're only looking for one probability value from this distribution, so we just need the probability of the given number of matches.

```python
class SuperTinder(Tinder):
    def Likelihood(self, data, hypo):
        """
        Compute the likelihood of 'data' given 'hypo'.

        data: tuple consisting of the number of matches and a boolean for whether there was a super like
        hypo: tuple of the total number of swipes and the probability of matching

        returns: likelihood of the data given the probability of matching and the total number of swipes
        """
        swipes, matchProb = hypo
        matches, superLike = data
        return 0 if matches > swipes else (MakeBinomialPmf(swipes, matchProb) + MakeBinomialPmf(1, matchProb * 3))[matches] if (superLike and matchProb * 3 <= 1) else binom.pmf(matches, swipes, matchProb)
```

Now we just need to update our new SuperTinder class with the data from before and after the super like.

```python
tSuper = SuperTinder(product(range(5, 201), np.linspace(0, 1, 101)))
for matches in [2, 4, 3, 2, 6, 4, 5]:
    tSuper.Update((matches, False))
for matches in [8, 2, 1]:
    tSuper.Update((matches, True))
thinkplot.Contour(tSuper, pcolor = True)
thinkplot.Config(xlabel = 'Right Swipes', ylabel = 'Matching Probability')
```

```python
thinkplot.Pdf(tSuper.Marginal(0))
thinkplot.Config(xlabel = 'Right Swipes', ylabel = 'PMF')
```

```python
thinkplot.Pdf(tSuper.Marginal(1))
thinkplot.Config(xlabel = 'Matching Probability', ylabel = 'PMF')
```

We have a distribution for Joe's swipe number and match rate under his belief that super likes are more effective than normal swiping, so now we just have to compare it to the distribution where super likes aren't treated any differently than swiping right.  To get this distribution, we need to create a new SuperTinder and update it with all of Joe's data without treating super likes differently.

```python
tNotSuper = SuperTinder(product(range(5, 201), np.linspace(0, 1, 101)))
for matches in [2, 4, 3, 2, 6, 4, 5, 8, 2, 1]:
    tNotSuper.Update((matches, False))
thinkplot.Contour(tNotSuper, pcolor = True)
thinkplot.Config(xlabel = 'Right Swipes', ylabel = 'Matching Probability')
```

```python
thinkplot.Pdf(tNotSuper.Marginal(0), label = 'Not Super')
thinkplot.Pdf(tSuper.Marginal(0), label = 'Super')
thinkplot.Config(xlabel = 'Right swipes', ylabel = 'PMF')
```

```python
thinkplot.Pdf(tNotSuper.Marginal(1), label = 'Not Super')
thinkplot.Pdf(tSuper.Marginal(1), label = 'Super')
thinkplot.Config(xlabel = 'Matching Probability', ylabel = 'PMF')
```

As shown in the plots, Joe's belief that super likes are thrice as effective as swiping right lowers the probabilies of lower values for total swipes, whereas treating super likes the same way we treat right swipes causes the distribution to be more normally curved.  Additionally, at larger total swipe values the marginal distribution plot for right swipes begins to exhibit some strange oscillating behavior that I am unable to explain, but it dies down by the time it reaches 600 swipes.
